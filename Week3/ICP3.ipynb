```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the CSV file into a DataFrame
# Make sure your CSV file is named 'Salary_Data.csv' and located in the same directory as the script
df = pd.read_csv('Salary_Data.csv')

# Print the first few rows to confirm data is loaded correctly
print(df.head())

# Split data into features (X) and target variable (y)
X = df[['YearsExperience']]  # Feature: Years of Experience
y = df['Salary']  # Target: Salary

# Split the data into training (2/3) and testing (1/3) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=42)

# Initialize and train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)  # Train the model on training data

# Make predictions on the test data
y_pred = model.predict(X_test)

# Calculate the Mean Squared Error (MSE) to evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

# Visualize the data and the regression line
plt.scatter(X_train, y_train, color='blue', label='Train Data')  # Plot training data
plt.scatter(X_test, y_test, color='green', label='Test Data')  # Plot test data
plt.plot(X, model.predict(X), color='red', label='Regression Line')  # Plot regression line

# Adding labels and title to the plot
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs. Experience')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step b: Read the provided CSV file 'data.csv'
# Make sure your CSV file is named 'data.csv' and located in the same directory as the script
df = pd.read_csv('data.csv')

# Step c: Show basic statistical description about the data
# This will give us the mean, min, max, standard deviation, etc. for all numerical columns
print("\nBasic Statistical Description of the Data:")
print(df.describe())

# Step d: Check if the data has null values
# This checks for any missing (null) values in the dataset
print("\nChecking for Null Values:")
print(df.isnull().sum())

# Step i: Replace the null values with the mean
# We replace null values in the DataFrame with the mean of each respective column
df.fillna(df.mean(), inplace=True)

# Step e: Select at least two columns and aggregate the data using min, max, count, mean
# For example, we'll select 'Duration' and 'Calories' columns and perform the aggregation
agg_result = df[['Duration', 'Calories']].agg(['min', 'max', 'count', 'mean'])
print("\nAggregation Results for 'Duration' and 'Calories':")
print(agg_result)

# Step f: Filter the DataFrame to select the rows with Calories values between 500 and 1000
calories_filtered = df[(df['Calories'] >= 500) & (df['Calories'] <= 1000)]
print("\nRows with Calories between 500 and 1000:")
print(calories_filtered)

# Step g: Filter the DataFrame to select the rows with Calories > 500 and Pulse < 100
filtered_condition = df[(df['Calories'] > 500) & (df['Pulse'] < 100)]
print("\nRows with Calories > 500 and Pulse < 100:")
print(filtered_condition)

# Step h: Create a new df_modified DataFrame that contains all columns except for Maxpulse
# We'll create a new DataFrame that does not contain the 'Maxpulse' column
df_modified = df.drop(columns=['Maxpulse'])
print("\nDataFrame with 'Maxpulse' column removed (df_modified):")
print(df_modified.head())

# Step i: Delete the Maxpulse column from the main df DataFrame
# This will remove the 'Maxpulse' column directly from the original DataFrame (df)
df.drop(columns=['Maxpulse'], inplace=True)
print("\nDataFrame with 'Maxpulse' column deleted:")
print(df.head())

# Step j: Convert the datatype of the Calories column to int datatype
# We'll convert the 'Calories' column to integer type using astype()
df['Calories'] = df['Calories'].astype(int)
print("\nDataFrame with 'Calories' column converted to integer:")
print(df.head())

# Step k: Using pandas, create a scatter plot for the two columns 'Duration' and 'Calories'
# We'll plot a scatter plot to visualize the relationship between 'Duration' and 'Calories'
plt.scatter(df['Duration'], df['Calories'], color='blue', alpha=0.5)
plt.title('Scatter Plot of Duration vs Calories')
plt.xlabel('Duration (minutes)')
plt.ylabel('Calories')
plt.show()
```
